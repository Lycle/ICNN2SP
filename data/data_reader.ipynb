```python
import pickle
```

```python
file_path = 'results.pkl'

with open(file_path, 'rb') as file:
    # Load only the top-level structure
    data = pickle.load(file)
    print(f"Type of data: {type(data)}")

    # If it's a dictionary, print the keys
    if isinstance(data, dict):
        print("Keys:", list(data.keys()))

    # If it's a list, print the length and first few items
    elif isinstance(data, list):
        print("List length:", len(data))
        print("First few items:", data[:5])

    # For other data types, print a summary
    else:
        print("Preview:", str(data)[:500])  # Show the first 500 characters
```

```python
# Explore 'data' key
if 'data' in data:
    print("Type of 'data':", type(data['data']))
    if isinstance(data['data'], dict):
        print("Keys in 'data':", list(data['data'].keys()))
    elif isinstance(data['data'], list):
        print("'data' list length:", len(data['data']))
        print("First few items in 'data':", data['data'][2])
    else:
        print("'data' preview:", str(data['data'])[2])
```
